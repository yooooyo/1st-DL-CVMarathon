```python
import requests
import bs4
import re
import time
import threading

class av:
    def __init__(self,rowData):
        for a in rowData.find_all('a'):
            if a.string:
                self.topic = a.string
#                 self.fnumber = re.
            if 'magnet' in a.get('href'):
                self.magnet = a.get('href')
        for td in rowData.find_all('td'):
            if td.get('data-timestamp'):
                self.size = td.previous_sibling.previous_sibling.string
                self.timestamp = td.string

getPage = lambda search='',page='': bs4.BeautifulSoup(requests.get(f'https://sukebei.nyaa.si/?f=2&c=2_2&q={search}&p={page}').text,'html')
getRowData = lambda data:[tr for tr in data.find_all('tr') if tr.get('class')]
getPagedata = lambda search,page: getRowData(getPage(search,page))

def crawl(search=[],page=0):
    def parsePage(s,p):
        page_data = []
        for _p in range(p+1):
            rowdata = getPagedata(s,_p)
            if rowdata:
                for row in rowdata:
                    page_data.append(av(row))
            if len(rowdata)<75:
                break
            time.sleep(1)
        return page_data
    if not search:
        data = parsePage('',page)
    for s in search:
        data = parsePage(s,page)
    return data

# search = ['abp','ssni']
# threads = []
# for i in search:
#     threads.append(threading.Thread(target=crawl,args=(i,0)))
#     threads[i].start()
```

```python
import time
import threading
import queue

# Worker 類別，負責處理資料
class Worker(threading.Thread):
  def __init__(self, queue, num):
    threading.Thread.__init__(self)
    self.queue = queue
    self.num = num

  def run(self):
    while self.queue.qsize() > 0:
      # 取得新的資料
      msg = self.queue.get()

      # 處理資料
      print("Worker %d: " % (self.num))
      for a in msg:
        print(a.topic)
      time.sleep(1)

# 建立佇列
my_queue = queue.Queue()

# 將資料放入佇列
# for i in range(10):
#   my_queue.put("Data %d" % i)
for i in ['ssni','abp','abw','fsdss','ipx']:
    my_queue.put(crawl([i],0))


# 建立兩個 Worker
my_worker1 = Worker(my_queue, 1)
my_worker2 = Worker(my_queue, 2)

# 讓 Worker 開始處理資料
my_worker1.start()
my_worker2.start()

# 等待所有 Worker 結束
my_worker1.join()
my_worker2.join()


print("Done.")
```
